```python
import pandas as pd
import datetime
import matplotlib.pyplot as plt
import holidays

veri=pd.read_excel("C:/Users/HP/Documents/Serpil Veri Bilimi/Zaman Serileri/EVDSDöviz.xlsx")
veri2=pd.read_excel("C:/Users/HP/Documents/Serpil Veri Bilimi/Zaman Serileri/EVDSDövizNan.xlsx")
veri2.drop("Unnamed: 2",axis=1,inplace=True)
veri2

print(veri.isnull().sum())
print(veri2.isnull().sum())
veri2.dtypes
veri2.drop(veri2.index[1546::],inplace=True)
veri2
```

```python
Tarihler=[]
Gunler=[]

for i in veri2["Tarih"]:
    tarihObje=datetime.datetime.strptime(i,"%d-%m-%Y")
    Tarihler.append(tarihObje)
    saatDilimi=datetime.timezone(datetime.timedelta(hours=3))
    tarihObje=tarihObje.astimezone(saatDilimi)
    gun=tarihObje.strftime("%A")
    Gunler.append(gun)

veri2["Tarih"]=Tarihler
veri2["Günler"]=Gunler
veri2=veri2[["Tarih", "Günler", "USD SATIŞ"]]
veri2
#veri.set_index("Tarih", inplace=True)
#veri.sort_values(by="Tarih",inplace=True)
```
